```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('311_Service_Requests_2020.csv')
```

```python
# initial dataframe
df
```

```python
# df['Incident Zip'].dtype
```

```python
# Chosen ZIP code: 10025 
my_zip = float(10025)
```

```python
df_myzip = df[df['Incident Zip'] == my_zip]
df_myzip
```

```python
parking_my = df_myzip.groupby('Complaint Type').size()['Illegal Parking']/df_myzip.groupby('Complaint Type').size().sum()
```

```python
parking_total = df.groupby('Complaint Type').size()['Illegal Parking']/df.groupby('Complaint Type').size().sum()
```

```python
if parking_my > parking_total:
    higher_parking_proportion = True
else:
    higher_parking_proportion = False
# higher_parking_proportion
```
